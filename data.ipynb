```python
df['length1'] = df['length1'].replace(1, 'short')
df['length2'] = df['length2'].replace(1, 'short')
df['length3'] = df['length3'].replace(1, 'short')
df['length4'] = df['length4'].replace(1, 'short')
df['length5'] = df['length5'].replace(1, 'short')

df['length1'] = df['length1'].replace(2, 'long')
df['length2'] = df['length2'].replace(2, 'long')
df['length3'] = df['length3'].replace(2, 'long')
df['length4'] = df['length4'].replace(2, 'long')
df['length5'] = df['length5'].replace(2, 'long')

df['length1'] = df['length1'].replace(3, 'long')
df['length2'] = df['length2'].replace(3, 'long')
df['length3'] = df['length3'].replace(3, 'long')
df['length4'] = df['length4'].replace(3, 'long')
df['length5'] = df['length5'].replace(3, 'long')
```

```python
colunas = ['shape1','shape2','shape3','shape4','shape5']

mapeamento = {
    1: 'engine',
    2: 'closedrect',
    3: 'openrect',
    4: 'dblclosedrect',
    5: 'dblopnrect',
    6: 'ellipse',
    7: 'hexagon',
    8: 'jaggedtop',
    9: 'closedtrap',
    10: 'opentrap',
    11: 'slopetoprect',
    12: 'slopetoptrap',
    13: 'closedushaped',
    14: 'openushaped',
    15: 'dblslopetoprect'
}

for coluna in colunas:
    df[coluna] = df[coluna].replace(mapeamento)
```

```python
colunas = ['load_shape1','load_shape2','load_shape3','load_shape4','load_shape5']

map_load = {
    1: "circlelod",
    2: "hexagonlod",
    3: "rectanglod",
    4: "trianglod"
}

for coluna in colunas:
    df[coluna] = df[coluna].replace(map_load)
```

```python
df.iloc[0:24 + 1, df.columns.get_loc('Class_attribute')] = 'east'
df.iloc[25:49 + 1, df.columns.get_loc('Class_attribute')] = 'west'

df.iloc[50:74 + 1, df.columns.get_loc('Class_attribute')] = 'east'
df.iloc[75:99 + 1, df.columns.get_loc('Class_attribute')] = 'west'
```

```python
# Função para verificar se a condição é satisfeita
def check_conditions(row):
    conditions = {
        'Rectangle_next_to_rectangle': 0,
        'Rectangle_next_to_triangle': 0,
        'Rectangle_next_to_hexagon': 0,
        'Rectangle_next_to_circle': 0,
        'Triangle_next_to_triangle': 0,
        'Triangle_next_to_hexagon': 0,
        'Triangle_next_to_circle': 0,
        'Hexagon_next_to_hexagon': 0,
        'Hexagon_next_to_circle': 0,
        'Circle_next_to_circle': 0
    }
    
    shapes = row[['load_shape1', 'load_shape2', 'load_shape3', 'load_shape4', 'load_shape5']].values
    
    for i in range(len(shapes) - 1):
        current_shape = shapes[i]
        next_shape = shapes[i + 1]
        
        if current_shape == 'rectanglod':  # Rectangle
            if next_shape == 'rectanglod':
                conditions['Rectangle_next_to_rectangle'] = 1
            elif next_shape == 'trianglod':
                conditions['Rectangle_next_to_triangle'] = 1
            elif next_shape == 'hexagonlod':
                conditions['Rectangle_next_to_hexagon'] = 1
            elif next_shape == 'circlelod':
                conditions['Rectangle_next_to_circle'] = 1
        elif current_shape == 'trianglod':  # Triangle
            if next_shape == 'trianglod':
                conditions['Triangle_next_to_triangle'] = 1
            elif next_shape == 'hexagonlod':
                conditions['Triangle_next_to_hexagon'] = 1
            elif next_shape == 'circlelod':
                conditions['Triangle_next_to_circle'] = 1
        elif current_shape == 'hexagonlod':  # Hexagon
            if next_shape == 'hexagonlod':
                conditions['Hexagon_next_to_hexagon'] = 1
            elif next_shape == 'circlelod':
                conditions['Hexagon_next_to_circle'] = 1
        elif current_shape == 'circlelod':  # Circle
            if next_shape == 'circlelod':
                conditions['Circle_next_to_circle'] = 1

    return pd.Series(conditions)

# Aplicar a função ao dataframe
df_conditions = df.apply(check_conditions, axis=1)

# Concatenar as novas colunas ao dataframe original
df = pd.concat([df, df_conditions], axis=1)

print(df)
```

```python
df.to_csv("trains-100-mod.csv")
```
